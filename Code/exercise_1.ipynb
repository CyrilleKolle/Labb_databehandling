#### using plotly


```python
import plotly_express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# inpired from
# https://stackoverflow.com/questions/71900635/how-to-create-subplots-in-plotly-express

# https://plotly.com/python/subplots/

"""Its not possible to plot subplots with plotly_express from my understanding so far, but using make_plots from plotly.subplots can assist in doing so by adding data from px.line to the grap
graph_object
"""
subplot_titles = ['Total death per week', 'Total cases per week',
                  'Cummulative cases per week', 'Relationship between deaths and new cases per week']
fig = make_subplots(rows=2, cols=2, row_heights=[
                    3, 3], column_widths=[3, 3], subplot_titles=subplot_titles, vertical_spacing=(0.1))
trace1 = px.line(weekly_data_kingdom, x='week',
                 y='Antal_avlidna_vecka', log_y=True)
trace2 = px.line(weekly_data_kingdom, x='week',
                 y='Antal_fall_vecka', log_y=True)
trace3 = px.line(weekly_data_kingdom, x='week', y='Kum_antal_fall')


trace_list = [trace1, trace2]
y_col = ["Antal_fall_vecka", "Antal_avlidna_vecka"]


# How to solve the scale of the y-axis
# https://stackoverflow.com/questions/61041707/plotly-log-scale-in-subplot-python

fig.add_trace(go.Scatter(name='Antal_avlidna_vecka', line={'dash': 'solid', 'color': 'orange'},
                         mode='lines', x=trace1.data[0]['x'], y=trace1.data[0]['y']), row=1, col=1)
fig.add_trace(go.Scatter(name='Antal_fall_vecka', line={'dash': 'solid', 'color': 'orange'},
                         mode='lines', x=trace2.data[0]['x'], y=trace2.data[0]['y']), row=1, col=2)
fig.add_trace(go.Scatter(name='Kum_antal_fall', line={'dash': 'solid', 'color': 'orange'},
                         mode='lines', x=trace3.data[0]['x'], y=trace3.data[0]['y']), row=2, col=1)

lines = [{'dash': 'solid', 'color': 'green'},
         {'dash': 'dash', 'color': 'blue'}]


for i, (item, title) in enumerate(zip(trace_list, y_col)):
    fig.add_trace(go.Scatter(line=lines[i], name=title,
                             mode='lines', x=item.data[0]['x'], y=item.data[0]['y'], showlegend=True), row=2, col=2)

# guide from:
# https://plotly.com/python/reference/layout/xaxis/
fig['layout'].update(height=1000, width=1200,
                     plot_bgcolor="rgba(255,255,255,0.1)",paper_bgcolor='rgba(255,255,255,0.9)', showlegend=False)
fig.update_yaxes(type='log', tickmode='auto', nticks=7, showgrid=False,
                 zeroline=True, linecolor="#000", showline=True, spikecolor="#000000")
fig.update_xaxes(tickmode='auto', nticks=5, showgrid=False, zeroline=True,
                 linecolor="#000", showline=True, spikecolor="#000000", showticklabels=True)


fig.layout["xaxis1"].title.text = "week"
fig.layout["xaxis2"].title.text = "week"
fig.layout["xaxis3"].title.text = "week"
fig.layout["xaxis4"].title.text = "week"

fig.layout["yaxis1"].title.text = "total deaths per week"
fig.layout["yaxis2"].title.text = "total cases per week"
fig.layout["yaxis3"].title.text = "cummulative cases per week"
fig.layout["yaxis4"].title.text = "total deaths/new cases per week"

fig.show()
```
